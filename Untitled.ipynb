```python
import psycopg
from psycopg.errors import DuplicateDatabase

# 1. Łączymy się z domyślną bazą systemową 'postgres'
konfiguracja = {
    "dbname": "postgres",  
    "user": "student15",
    "password": "fwbmfhgxqdZNP",
    "host": "localhost",
    "port": "5432"
}

try:
    # UWAGA: Tworzenie bazy wymaga włączenia trybu autocommit=True
    with psycopg.connect(**konfiguracja, autocommit=True) as conn:
        with conn.cursor() as cur:
            # Tworzymy naszą nową bazę o nazwie 'biblioteka'
            cur.execute("CREATE DATABASE biblioteka;")
            print("Sukces! Baza 'biblioteka' została utworzona.")

except DuplicateDatabase:
    print("Baza 'biblioteka' już istnieje na tym serwerze.")
except Exception as e:
    print(f"Coś poszło nie tak: {e}")
```
